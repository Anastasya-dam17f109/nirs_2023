```python
def drawResult():
    image   = "D:\\generated_image.txt";
    spl_mix = "D:\\splitted_image.txt";
    gridsize = (1, 2)
    fig = plt.figure(figsize=(18, 16))
    ax1 = plt.subplot2grid(gridsize, (0, 0))
    ax2 = plt.subplot2grid(gridsize, (0, 1))
    data = np.loadtxt(image)
    pc = ax1.contourf(data)
    plt.colorbar(pc, ax=ax1, format='$%d')
    data1 = np.loadtxt(spl_mix)
    pc1 = ax2.contourf(data1)
    plt.colorbar(pc1, ax=ax2, format='$%d')
    plt.show()  
```

```python
from PIL import Image
from PIL import ImageDraw

def drawResultInFile(filename):
    image   = "D:\\generated_image.txt";
    spl_mix = "D:\\splitted_image.txt";
    data1 = np.loadtxt(spl_mix)
    width, height= data1.shape
    #width, height = height, width
    img = Image.new('RGB', ( height, width), color=0)
    draw  = ImageDraw.Draw(img)
    t_color = ['blue', 'red', 'yellow', 'green' , 'violet']
    
    width, height = data1.shape
    for i in range(width):
        for j in range(height):
            draw.point((j,width - i - 1), fill=t_color[int(data1[i,j])-1])
    img.save(filename)
    img.close()
    
```
