```python
import math
amount = 3

attr = []

with open(results) as file:
    for i in file:
        f, s = i.split(":")
        #print(s)
        attr.append(float(s))
print(attr)
cl1 = 0
cl2 = 0
cl3 = 0
cl1_1 = 0
cl2_1 = 0
cl3_1 = 0
for j in range (amount):
    cl1 = 0
    cl2 = 0
    cl3 = 0
    cl1_1 = 0
    cl2_1 = 0
    cl3_1 = 0
    for i in range(epocs):
        cl1 += attr[j*3*epocs + 3*i]
        cl2 += attr[j*3*epocs +3*i +1]
        cl3 += attr[j*3*epocs +3*i +2]
    cl1 /= epocs
    cl2 /= epocs
    cl3 /= epocs
    print("class mids ", cl1,cl2, cl3)
    for i in range(epocs):
        cl1_1 += (attr[j*3*epocs + 3*i] - cl1 )*(attr[j*3*epocs + 3*i] - cl1 )
        cl2_1 += (attr[j*3*epocs +3*i +1] - cl2 )*(attr[j*3*epocs +3*i +1] - cl2 )
        cl3_1 += (attr[j*3*epocs +3*i +2] - cl3)*(attr[j*3*epocs +3*i +2] - cl3)
    cl1_1 /= epocs
    cl2_1 /= epocs
    cl3_1 /= epocs
    cl1_1 = math.sqrt(cl1_1)
    cl2_1 = math.sqrt(cl2_1)
    cl3_1 = math.sqrt(cl3_1)
    print("class diffs ", cl1_1,cl2_1, cl3_1)
```
